```python
clean_data_01 = df.drop(columns=["ORIGIN_CITY_NAME", "DEST_CITY_NAME"])

print("Size of new dataset:", clean_data_01.shape)

clean_data_01.to_csv("clean_data_01.csv", index=False)
```

```python
# Delete rows with CANCELLED = 1
clean_data_01 = clean_data_01[clean_data_01["CANCELLED"] != 1]

print("Size of dataset:", clean_data_01.shape)

clean_data_01.to_csv("clean_data_01.csv", index=False)
```

```python
total_null = clean_data_01.isnull().sum().sum()
print("Total NULL value in dataset:", total_null)

null_per_column = clean_data_01.isnull().sum()
print("\nNULL per collums:\n", null_per_column[null_per_column > 0])
```

```python
clean_data_01 = clean_data_01.dropna()
clean_data_01.to_csv("clean_data_01.csv", index=False)
clean_data_01.shape
```

```python
def hhmm_to_decimal(time_val):
    """
    Convert time from HHMM (e.g., 1130, 845) to decimal hour (e.g., 11.5, 8.75).
    """
    if pd.isna(time_val):
        return None
    
    time_val = int(time_val)
    hour = time_val // 100
    minute = time_val % 100
    decimal_time = hour + minute/60
    return round(decimal_time, 2)


clean_data_02 = clean_data_01.copy()
clean_data_02["CRS_DEP_TIME"] = clean_data_02["CRS_DEP_TIME"].apply(hhmm_to_decimal)
clean_data_02["CRS_ARR_TIME"] = clean_data_02["CRS_ARR_TIME"].apply(hhmm_to_decimal)

clean_data_02.to_csv("clean_data_02.csv", index=False)
```
