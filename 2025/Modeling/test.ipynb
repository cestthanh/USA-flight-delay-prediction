```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\Project DS Final\check\flight_dep_linear.csv')
```

```python
# Lọc các cột cần loại bỏ
drop_cols = [col for col in df.columns if col.startswith('DEST_') or col.startswith('ORIGIN_')]

# Tạo DataFrame mới sau khi loại bỏ các cột
df_reduced = df.drop(columns=drop_cols)

# Xuất DataFrame đã giảm cột sang file CSV mới
df_reduced.to_csv("test01.csv", index=False)

print("✅ DataFrame đã được lưu vào file 'test01.csv'")
```
