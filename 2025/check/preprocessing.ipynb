```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("final_clean_data.csv")
df.head()
```

```python
keep_cols = [
    "MONTH", "DAY_OF_MONTH", "DAY_OF_WEEK",
    "CRS_DEP_TIME", "CRS_ELAPSED_TIME",
    "OP_UNIQUE_CARRIER", "ORIGIN", "DEST",
    "DISTANCE",
    "HourlyDewPointTemperature", "HourlyDryBulbTemperature",
    "HourlyRelativeHumidity", "HourlyVisibility", "HourlyWindSpeed",
    "DEP_DELAY"
]

flight_dep_model = df[keep_cols].copy()
print("‚úÖ Dataset cho m√¥ h√¨nh delay c·∫•t c√°nh:", flight_dep_model.shape)
flight_dep_model.head(5)
```

### Encode CRS_DEP_TIME th√†nh sin/cos (chu k·ª≥ 24h)

```python
import numpy as np

# --- CRS_DEP_TIME ƒë√£ ·ªü d·∫°ng float, v√≠ d·ª•: 6.5 = 6h30, 13.75 = 13h45 ---

# Encode cyclical (chu k·ª≥ 24h)
flight_dep_model["DEP_HOUR_sin"] = np.sin(2 * np.pi * flight_dep_model["CRS_DEP_TIME"] / 24)
flight_dep_model["DEP_HOUR_cos"] = np.cos(2 * np.pi * flight_dep_model["CRS_DEP_TIME"] / 24)

# (T√πy ch·ªçn) B·ªè c·ªôt g·ªëc n·∫øu kh√¥ng d√πng n·ªØa
# flight_dep_model.drop(columns=["CRS_DEP_TIME"], inplace=True)

# Ki·ªÉm tra v√†i d√≤ng ƒë·∫ßu
print(flight_dep_model[["CRS_DEP_TIME", "DEP_HOUR_sin", "DEP_HOUR_cos"]].head(10))
```

```python
# --- 3Ô∏è‚É£ X√°c ƒë·ªãnh categorical v√† numeric features ---
categorical_cols = ["MONTH", "DAY_OF_WEEK", "OP_UNIQUE_CARRIER", "ORIGIN", "DEST"]

numeric_cols = [
    "DAY_OF_MONTH", "CRS_ELAPSED_TIME", "DISTANCE",
    "HourlyDewPointTemperature", "HourlyDryBulbTemperature",
    "HourlyRelativeHumidity", "HourlyVisibility", "HourlyWindSpeed",
    "DEP_HOUR_sin", "DEP_HOUR_cos"
]
```

### Encoded cho ph·∫ßn linear, NN, ...

##### Encoded 5 features categorical b·∫±ng one-hot

```python
# --- 4Ô∏è‚É£ One-Hot Encode b·∫±ng sklearn ---
ohe = OneHotEncoder(
    sparse_output=False,   # Xu·∫•t DataFrame d·∫°ng ƒë·∫ßy ƒë·ªß thay v√¨ ma tr·∫≠n th∆∞a
    drop=None,             # Gi·ªØ t·∫•t c·∫£ c√°c c·ªôt (kh√¥ng drop_first)
    handle_unknown="ignore"  # B·ªè qua category m·ªõi trong t·∫≠p test
)

# Fit v√† transform categorical features
encoded_array = ohe.fit_transform(flight_dep_model[categorical_cols])

# L·∫•y t√™n c·ªôt m·ªõi
encoded_cols = ohe.get_feature_names_out(categorical_cols)

# T·∫°o DataFrame ch·ª©a c√°c c·ªôt one-hot
encoded_df = pd.DataFrame(encoded_array, columns=encoded_cols, index=flight_dep_model.index)


# --- 5Ô∏è‚É£ G·ªôp numeric + one-hot + target ---
flight_dep_onehot = pd.concat(
    [flight_dep_model[numeric_cols], encoded_df, flight_dep_model[["DEP_DELAY"]]],
    axis=1
)

# --- 6Ô∏è‚É£ Ki·ªÉm tra k·∫øt qu·∫£ ---
print("\n‚úÖ Dataset sau One-Hot Encoding:")
print("S·ªë d√≤ng:", flight_dep_onehot.shape[0])
print("S·ªë c·ªôt:", flight_dep_onehot.shape[1])

# Hi·ªÉn th·ªã v√†i c·ªôt m·ªõi ƒë·ªÉ ki·ªÉm tra
print("\nüß± M·ªôt v√†i c·ªôt m·ªõi ƒë∆∞·ª£c t·∫°o:")
print(flight_dep_onehot[encoded_cols[:10]].head(3))

# --- 7Ô∏è‚É£ (T√πy ch·ªçn) L∆∞u dataset ƒë√£ encode ƒë·ªÉ d√πng cho Linear/NN ---
flight_dep_onehot.to_csv("flight_dep_linear.csv", index=False)
```

### Encode cho ph·∫ßn Tree-based (RF, XGB, LGBM)



#### Label encoded cho OP_UNIQUE_CARRIER, ORIGIN, DEST
#### C√≤n MONTH, DAY_OF_WEEK ordinal encoded (v·∫´n nh∆∞ v·∫≠y)

```python
label_cols = ["OP_UNIQUE_CARRIER", "ORIGIN", "DEST"]
label_encoders = {}

for col in label_cols:
    le = LabelEncoder()
    flight_dep_model[col] = le.fit_transform(flight_dep_model[col])
    label_encoders[col] = le
    print(f"üî§ {col}: {len(le.classes_)} categories encoded.")


# --- 4Ô∏è‚É£ Gi·ªØ l·∫°i c√°c c·ªôt numeric + target ---
tree_ready_cols = [
    "MONTH", "DAY_OF_MONTH", "DAY_OF_WEEK",
    "CRS_ELAPSED_TIME", "DISTANCE",
    "HourlyDewPointTemperature", "HourlyDryBulbTemperature",
    "HourlyRelativeHumidity", "HourlyVisibility", "HourlyWindSpeed",
    "DEP_HOUR_sin", "DEP_HOUR_cos",
    "OP_UNIQUE_CARRIER", "ORIGIN", "DEST",
    "DEP_DELAY"
]

flight_dep_tree = flight_dep_model[tree_ready_cols]


# --- 5Ô∏è‚É£ Ki·ªÉm tra t·ªïng quan ---
print("\n‚úÖ Dataset sau khi encode:")
print("S·ªë d√≤ng:", flight_dep_tree.shape[0])
print("S·ªë c·ªôt:", flight_dep_tree.shape[1])
print("\nüìä Ki·ªÉu d·ªØ li·ªáu:")
print(flight_dep_tree.dtypes.value_counts())

print("\nüîç Xem v√†i d√≤ng ƒë·∫ßu:")
print(flight_dep_tree.head(5))


# --- 6Ô∏è‚É£ (T√πy ch·ªçn) L∆∞u l·∫°i ƒë·ªÉ train model ---
flight_dep_tree.to_csv("flight_dep_tree.csv", index=False)
```
