```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("final_clean_data.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
keep_cols = [
    "MONTH", "DAY_OF_MONTH", "DAY_OF_WEEK",
    "CRS_DEP_TIME", "CRS_ELAPSED_TIME",
    "OP_UNIQUE_CARRIER", "ORIGIN", "DEST",
    "DISTANCE",
    "HourlyDewPointTemperature", "HourlyDryBulbTemperature",
    "HourlyRelativeHumidity", "HourlyVisibility", "HourlyWindSpeed",
    "DEP_DELAY"
]

flight_dep_model = df[keep_cols].copy()
print("✅ Dataset cho mô hình delay cất cánh:", flight_dep_model.shape)
flight_dep_model.head(5)
```

### Encode CRS_DEP_TIME thành sin/cos (chu kỳ 24h)

```python
import numpy as np

# --- CRS_DEP_TIME đã ở dạng float, ví dụ: 6.5 = 6h30, 13.75 = 13h45 ---

# Encode cyclical (chu kỳ 24h)
flight_dep_model["DEP_HOUR_sin"] = np.sin(2 * np.pi * flight_dep_model["CRS_DEP_TIME"] / 24)
flight_dep_model["DEP_HOUR_cos"] = np.cos(2 * np.pi * flight_dep_model["CRS_DEP_TIME"] / 24)

# (Tùy chọn) Bỏ cột gốc nếu không dùng nữa
# flight_dep_model.drop(columns=["CRS_DEP_TIME"], inplace=True)

# Kiểm tra vài dòng đầu
print(flight_dep_model[["CRS_DEP_TIME", "DEP_HOUR_sin", "DEP_HOUR_cos"]].head(10))
```

```python
# --- 3️⃣ Xác định categorical và numeric features ---
categorical_cols = ["MONTH", "DAY_OF_WEEK", "OP_UNIQUE_CARRIER", "ORIGIN", "DEST"]

numeric_cols = [
    "DAY_OF_MONTH", "CRS_ELAPSED_TIME", "DISTANCE",
    "HourlyDewPointTemperature", "HourlyDryBulbTemperature",
    "HourlyRelativeHumidity", "HourlyVisibility", "HourlyWindSpeed",
    "DEP_HOUR_sin", "DEP_HOUR_cos"
]
```

### Encoded cho phần linear, NN, ...

##### Encoded 5 features categorical bằng one-hot

```python
# --- 4️⃣ One-Hot Encode bằng sklearn ---
ohe = OneHotEncoder(
    sparse_output=False,   # Xuất DataFrame dạng đầy đủ thay vì ma trận thưa
    drop=None,             # Giữ tất cả các cột (không drop_first)
    handle_unknown="ignore"  # Bỏ qua category mới trong tập test
)

# Fit và transform categorical features
encoded_array = ohe.fit_transform(flight_dep_model[categorical_cols])

# Lấy tên cột mới
encoded_cols = ohe.get_feature_names_out(categorical_cols)

# Tạo DataFrame chứa các cột one-hot
encoded_df = pd.DataFrame(encoded_array, columns=encoded_cols, index=flight_dep_model.index)


# --- 5️⃣ Gộp numeric + one-hot + target ---
flight_dep_onehot = pd.concat(
    [flight_dep_model[numeric_cols], encoded_df, flight_dep_model[["DEP_DELAY"]]],
    axis=1
)

# --- 6️⃣ Kiểm tra kết quả ---
print("\n✅ Dataset sau One-Hot Encoding:")
print("Số dòng:", flight_dep_onehot.shape[0])
print("Số cột:", flight_dep_onehot.shape[1])

# Hiển thị vài cột mới để kiểm tra
print("\n🧱 Một vài cột mới được tạo:")
print(flight_dep_onehot[encoded_cols[:10]].head(3))

# --- 7️⃣ (Tùy chọn) Lưu dataset đã encode để dùng cho Linear/NN ---
flight_dep_onehot.to_csv("flight_dep_linear.csv", index=False)
```

### Encode cho phần Tree-based (RF, XGB, LGBM)



#### Label encoded cho OP_UNIQUE_CARRIER, ORIGIN, DEST
#### Còn MONTH, DAY_OF_WEEK ordinal encoded (vẫn như vậy)

```python
label_cols = ["OP_UNIQUE_CARRIER", "ORIGIN", "DEST"]
label_encoders = {}

for col in label_cols:
    le = LabelEncoder()
    flight_dep_model[col] = le.fit_transform(flight_dep_model[col])
    label_encoders[col] = le
    print(f"🔤 {col}: {len(le.classes_)} categories encoded.")


# --- 4️⃣ Giữ lại các cột numeric + target ---
tree_ready_cols = [
    "MONTH", "DAY_OF_MONTH", "DAY_OF_WEEK",
    "CRS_ELAPSED_TIME", "DISTANCE",
    "HourlyDewPointTemperature", "HourlyDryBulbTemperature",
    "HourlyRelativeHumidity", "HourlyVisibility", "HourlyWindSpeed",
    "DEP_HOUR_sin", "DEP_HOUR_cos",
    "OP_UNIQUE_CARRIER", "ORIGIN", "DEST",
    "DEP_DELAY"
]

flight_dep_tree = flight_dep_model[tree_ready_cols]


# --- 5️⃣ Kiểm tra tổng quan ---
print("\n✅ Dataset sau khi encode:")
print("Số dòng:", flight_dep_tree.shape[0])
print("Số cột:", flight_dep_tree.shape[1])
print("\n📊 Kiểu dữ liệu:")
print(flight_dep_tree.dtypes.value_counts())

print("\n🔍 Xem vài dòng đầu:")
print(flight_dep_tree.head(5))


# --- 6️⃣ (Tùy chọn) Lưu lại để train model ---
flight_dep_tree.to_csv("flight_dep_tree.csv", index=False)
```
